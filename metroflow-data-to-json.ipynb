```python
import pandas as pd
```

```python
df = pd.read_excel('./seoulmetro2016.xlsx')
print(len(df))
df.head()
```

```python
# mon tue wed thur fri sat sun
week = ['fri', 'sat', 'sun', 'mon', 'tue', 'wed', 'thu' ]

day_list = []

for i in range(0,len(df)):
    day_list.append(week[((i // 550)%7)])
    
df.insert(1, column='요일', value=day_list)
```

```python
# 16-01-01 is friday
df.head(2)
```

```python
# 16-12-31 is satrrday
df.tail(2)
```

```python
# concat inm out
df_in = df[df['구분'] == '승차'].reset_index()
df_out = df[df['구분'] == '하차'].reset_index()

df_in_out = pd.concat([df_in.ix[:, 1:], df_out.ix[:,6:]], axis=1, join_axes=[df_in.index])
df_in_out.head(2)
```

```python
# sample_df
sample_df = df_in_out.ix[:5,:]
sample_df
```

```python
# making data sample
for index, row in sample_df.iterrows():
    print(row['역번호'], row['역명'].split('(')[0], row[6:26].values, row[27:].values)
```

```python
'''
data = [
    df.iterrows
        json = {    
            insert
        }
    append
]
'''

sample_data = []

for index, row in sample_df.iterrows():
    json_data = {}
    json_data['stationId'] = row['역번호']
    json_data['stationName'] = row['역명'].split('(')[0]
    json_data['flowIn'] = list(row[6:26])
    json_data['flowOut'] = list(row[27:])
    
    sample_data.append(json_data)

sample_data
```

```python
# making data

data = []

for index, row in df_in_out.iterrows():
    json_data = {}
    json_data['stationId'] = row['역번호']
    json_data['stationName'] = row['역명'].split('(')[0]
    json_data['flowIn'] = list(row[6:26])
    json_data['flowOut'] = list(row[27:])
    
    data.append(json_data)

len(data)
```

```python
data[0]
```

```python
# making date type is "2016-01-01-fri"

def join_date(x):
    return x['날짜'] + '-' + x['요일']

dates = df_in_out.apply(join_date, axis=1)
date_list = list(sorted(list(set(dates))))
```

```python
# number of date
len(date_list)
```

```python
# number of data
len(data)
```

```python
# number of stations
len(data) / len(date_list)
```

```python
# making json_data
'''
{
    date_list loop
        data() loop
}
'''

json_data = []

for idx, val in enumerate(date_list):
    tmp_json = {}
    tmp_json['date'] = val
    tmp_json['data'] = data[idx*275:(idx+1)*275]
    
    json_data.append(tmp_json)
```

```python
# sample data to json
import json
with open('sample.json', 'w', encoding='utf-8') as fp:
    json.dump(sample_data, fp, ensure_ascii=False)
```

```python
# original data to json
import json
with open('metroflow.json', 'w', encoding='utf-8') as fp:
    json.dump(json_data, fp, ensure_ascii=False)
```
